### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
# Step 3: Clean the 'DatabaseWantToWorkWith' column and split entries
# We assume that the databases are separated by a delimiter like a semicolon

# First, let's split the databases into separate entries
databases = df['DatabaseWantToWorkWith'].dropna().str.split(';', expand=True).stack()

# Now, count the occurrences of each database
database_counts = databases.value_counts().head(5)

# Step 4: Create a pie chart for the top 5 most-desired databases
plt.figure(figsize=(8, 8))
plt.pie(database_counts, labels=database_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Most Desired Databases to Work With')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
#Clean the 'DevType' column and split entries
# We'll split the 'DevType' entries based on semicolons (assuming multiple roles are separated by a semicolon)

dev_types = df['DevType'].dropna().str.split(';', expand=True).stack()

# Count the occurrences of each developer type
dev_type_counts = dev_types.value_counts().head(5)

# Create a pie chart for the top 5 most common developer types
plt.figure(figsize=(8, 8))
plt.pie(dev_type_counts, labels=dev_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Developer Roles')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
# Step 1: Clean the 'NEWSOSites' column and split entries
# We assume that multiple operating systems are separated by a semicolon (like in the other columns)

opsys = df['NEWSOSites'].dropna().str.split(';', expand=True).stack()

# Step 2: Count the occurrences of each operating system
opsys_counts = opsys.value_counts().head(5)  # Get the top 5 operating systems

# Step 3: Create a pie chart for the top operating systems used for professional work
plt.figure(figsize=(8, 8))
plt.pie(opsys_counts, labels=opsys_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Top Operating Systems Used Professionally')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With
