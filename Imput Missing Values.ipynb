### Step 1: Import Required Libraries


```python
import pandas as pd
```

### Step 2: Load the Dataset Into a Dataframe


#### **Read Data**
<p>
The functions below will download the dataset into your browser:
</p>


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### Step 3. Finding and Removing Duplicates
##### Task 1: Identify duplicate rows in the dataset.


```python
## Write your code here
# Count duplicate rows
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")

# Display the first few duplicate rows
duplicates = df[df.duplicated()]
print("First few duplicate rows:")
print(duplicates.head())
```

##### Task 2: Remove the duplicate rows from the dataframe.



```python
## Write your code here
# Remove duplicate rows
df = df.drop_duplicates()

# Confirm removal by recounting duplicates
num_duplicates_after = df.duplicated().sum()
print(f"Number of duplicate rows after removal: {num_duplicates_after}")
```

### Step 4: Finding Missing Values
##### Task 3: Find the missing values for all columns.


```python
## Write your code here
# Count missing values in each column
missing_values = df.isnull().sum()

# Display columns with missing values
print("Missing values per column:")
print(missing_values[missing_values > 0])
```

##### Task 4: Find out how many rows are missing in the column RemoteWork.


```python
## Write your code here
# Count missing values in the 'RemoteWork' column
missing_remote_work = df['RemoteWork'].isnull().sum()

print(f"Number of missing values in 'RemoteWork' column: {missing_remote_work}")
```

### Step 5. Imputing Missing Values
##### Task 5: Find the value counts for the column RemoteWork.


```python
## Write your code here
# Get the value counts for the 'RemoteWork' column
remote_work_counts = df['RemoteWork'].value_counts()

print("Value counts for 'RemoteWork' column:")
print(remote_work_counts)
```

##### Task 6: Identify the most frequent (majority) value in the RemoteWork column.



```python
## Write your code here
# Find the most frequent value in the 'RemoteWork' column
most_frequent_remote_work = df['RemoteWork'].mode()[0]

print(f"The most frequent value in 'RemoteWork' column is: {most_frequent_remote_work}")
```

##### Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.



```python
# Impute missing values in 'RemoteWork' with the most frequent value
df['RemoteWork'] = df['RemoteWork'].fillna(most_frequent_remote_work)

# Verify if all missing values were imputed
missing_remote_work_after_imputation = df['RemoteWork'].isnull().sum()

print(f"Number of missing values in 'RemoteWork' column after imputation: {missing_remote_work_after_imputation}")
```

##### Task 8: Check for any compensation-related columns and describe their distribution.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check for compensation-related columns by inspecting the column names
compensation_columns = [col for col in df.columns if 'Comp' in col or 'Salary' in col]

# Display compensation-related columns if they exist
print("Compensation-related columns:")
print(compensation_columns)

# If any compensation-related column is found, describe its distribution
if compensation_columns:
    for col in compensation_columns:
        print(f"\nDistribution of {col}:")
        print(df[col].describe())
        
        # Visualize the distribution using a histogram
        plt.figure(figsize=(10, 6))
        sns.histplot(df[col], kde=True, bins=30)
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()
else:
    print("No compensation-related columns found.")
```
