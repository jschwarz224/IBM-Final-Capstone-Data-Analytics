```python
# Step 1: Download the dataset
!wget -O survey-data.csv https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv

# Step 2: Import necessary libraries and load the dataset
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("survey-data.csv")

# Display the first few rows to understand the structure of the data
df.head()
```

### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
## Write your code here
# Step 1: Drop NaN values from ConvertedCompYearly
comp_data = df['ConvertedCompYearly'].dropna()

# Step 2: Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(comp_data, bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Yearly Compensation')
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Number of Respondents')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Step 3: Show the plot
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
## Write your code here
# Step 1: Define a mapping from Age categories to approximate numeric values
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}

# Step 2: Apply the mapping
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Step 3: Drop NaN values
age_data = df['AgeNumeric'].dropna()

# Step 4: Plot the box plot
plt.figure(figsize=(8, 6))
plt.boxplot(age_data, vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Step 5: Show the plot
plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
## Write your code here
# Step 1: Prepare the data
scatter_df = df[['AgeNumeric', 'ConvertedCompYearly']].dropna()

# Step 2: Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(scatter_df['AgeNumeric'], scatter_df['ConvertedCompYearly'], alpha=0.5)
plt.title('Scatter Plot of Age vs. Yearly Compensation')
plt.xlabel('Age (Numeric)')
plt.ylabel('Yearly Compensation (USD)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Step 3: Show the plot
plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
## Write your code here
# Step 1: Prepare the data
bubble_df = df[['ConvertedCompYearly', 'JobSatPoints_6', 'AgeNumeric']].dropna()

# Step 2: Plot the bubble chart
plt.figure(figsize=(12, 7))
plt.scatter(
    bubble_df['ConvertedCompYearly'], 
    bubble_df['JobSatPoints_6'], 
    s=bubble_df['AgeNumeric']*5,  # Multiply to make bubbles visually larger
    alpha=0.5,
    edgecolors='w'
)
plt.title('Bubble Plot: Compensation vs Job Satisfaction (Bubble Size = Age)')
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Job Satisfaction Points (6)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Step 3: Show the plot
plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
## Write your code here
# Step 1: Prepare the data
mainbranch_counts = df['MainBranch'].value_counts()

# Step 2: Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Distribution of MainBranch (Primary Role)')
plt.xlabel('Number of Respondents')
plt.ylabel('Primary Role')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Step 3: Show the plot
plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
## Write your code here
# Step 1: Split the 'LanguageWantToWorkWith' column and stack the values
languages = df['LanguageWantToWorkWith'].dropna().str.split(';', expand=True).stack().reset_index(drop=True)

# Step 2: Count the frequency of each language
top_languages = languages.value_counts().head(5)

# Step 3: Plot the top 5 programming languages as a vertical bar chart
plt.figure(figsize=(10, 6))
top_languages.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 5 Programming Languages Respondents Want to Work With')
plt.xlabel('Programming Languages')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Step 4: Show the plot
plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
## Write your code here
# Step 1: Group by age and calculate the median of JobSatPoints_6 and JobSatPoints_7
age_grouped = df.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Step 2: Plot the stacked bar chart
age_grouped.plot(kind='bar', stacked=True, figsize=(10, 6), color=['lightblue', 'lightcoral'])

# Step 3: Add titles and labels
plt.title('Median Job Satisfaction (JobSatPoints_6 and JobSatPoints_7) by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Job Satisfaction')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Step 4: Show the plot
plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
## Write your code here
# Step 1: Split the DatabaseHaveWorkedWith column into individual databases
databases = df['DatabaseHaveWorkedWith'].dropna().str.split(';', expand=True).stack()

# Step 2: Count the frequency of each database
database_counts = databases.value_counts()

# Step 3: Plot the bar chart
database_counts.head(10).plot(kind='bar', figsize=(10, 6), color='lightgreen')

# Step 4: Add titles and labels
plt.title('Top 10 Most Popular Databases Used by Respondents')
plt.xlabel('Database')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Step 5: Show the plot
plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
## Write your code here
# Step 1: Group the data by Age and calculate the median ConvertedCompYearly
median_compensation_by_age = df.groupby('Age')['ConvertedCompYearly'].median()

# Step 2: Plot the grouped bar chart
median_compensation_by_age.plot(kind='bar', figsize=(10, 6), color='skyblue')

# Step 3: Add titles and labels
plt.title('Median Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Step 4: Show the plot
plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
## Write your code here
# Step 1: Group the data by Country and count the number of respondents
respondent_count_by_country = df['Country'].value_counts()

# Step 2: Plot the bar chart
respondent_count_by_country.plot(kind='bar', figsize=(12, 6), color='lightcoral')

# Step 3: Add titles and labels
plt.title('Distribution of Respondents by Country')
plt.xlabel('Country')
plt.ylabel('Respondent Count')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()

# Step 4: Show the plot
plt.show()
```
