### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Grouping the data by 'Age' and calculating the median of 'JobSatPoints_6' and 'JobSatPoints_7'
age_grouped = df.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Plotting the stacked bar chart
age_grouped.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#1f77b4', '#ff7f0e'])

# Adding labels and title
plt.title('Composition of Job Satisfaction Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Median Job Satisfaction Points')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Assuming 'LearnCode' and 'LearnCodeOnline' columns contain database mentions
# Combine relevant columns for database preferences
df['LearnCode'] = df['LearnCode'].fillna('')
df['LearnCodeOnline'] = df['LearnCodeOnline'].fillna('')

# Combine both columns into a single column
df['LearningPreferences'] = df['LearnCode'] + ';' + df['LearnCodeOnline']

# Split the learning preferences into separate entries (assuming semicolon as a separator)
learning_preferences = df['LearningPreferences'].str.split(';', expand=True).stack().reset_index(drop=True)

# Create a DataFrame with the 'Age' and the corresponding learning preferences
df_age_db = pd.concat([df['Age'], learning_preferences], axis=1)
df_age_db.columns = ['Age', 'PreferredDatabase']

# Optional: Filter out empty or irrelevant responses (e.g., "None", "No preference")
df_age_db = df_age_db[df_age_db['PreferredDatabase'].notna() & (df_age_db['PreferredDatabase'] != '')]

# Create a count of each database preference by age group
age_db_grouped = pd.crosstab(df_age_db['Age'], df_age_db['PreferredDatabase'])

# Plotting the stacked bar chart
age_db_grouped.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='tab20')

# Adding labels and title
plt.title('Top Databases Preferred by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Filter out rows where 'JobSat' is NaN
df_cleaned = df.dropna(subset=['JobSat'])

# Group the data by 'JobSat' (Job Satisfaction) and 'Employment' (Employment Type) to get counts
employment_job_sat = pd.crosstab(df_cleaned['JobSat'], df_cleaned['Employment'])

# Plotting the stacked bar chart
employment_job_sat.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='tab20')

# Adding labels and title
plt.title('Distribution of Employment Types by Job Satisfaction')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Filter out rows where 'JobSat' is NaN
df_cleaned = df.dropna(subset=['JobSat'])

# Handle multiple employment types by splitting into separate rows
df_employment_split = df_cleaned['Employment'].str.split(';', expand=True).stack().reset_index(drop=True)
df_cleaned_employment = pd.concat([df_cleaned[['JobSat']], df_employment_split], axis=1)
df_cleaned_employment.columns = ['JobSat', 'Employment']

# Now we can proceed with grouping by 'JobSat' and 'Employment'
employment_job_sat = pd.crosstab(df_cleaned_employment['JobSat'], df_cleaned_employment['Employment'])

# Plotting the stacked bar chart
employment_job_sat.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='tab20')

# Adding labels and title
plt.title('Distribution of Employment Types by Job Satisfaction')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Handle missing values in 'PlatformAdmired' and 'Employment' columns
df['PlatformAdmired'] = df['PlatformAdmired'].fillna('No preference')  # Fill NaN with 'No preference'
df['Employment'] = df['Employment'].fillna('No employment')  # Fill NaN with 'No employment'

# Split the 'PlatformAdmired' column by semicolons if there are multiple platforms listed
platform_preferences = df['PlatformAdmired'].str.split(';', expand=True).stack().reset_index(drop=True)

# Similarly, split the 'Employment' column by semicolons if multiple employment types are listed
employment_types = df['Employment'].str.split(';', expand=True).stack().reset_index(drop=True)

# Create a new DataFrame where each row represents one employment type and one platform admired
df_platform_employment = pd.concat([employment_types, platform_preferences], axis=1)
df_platform_employment.columns = ['Employment', 'AdmiredPlatform']

# Optional: Filter out 'No preference' if not relevant for the analysis
df_platform_employment = df_platform_employment[df_platform_employment['AdmiredPlatform'] != 'No preference']

# Group by 'Employment' and 'AdmiredPlatform' and count occurrences
employment_platform_grouped = pd.crosstab(df_platform_employment['Employment'], df_platform_employment['AdmiredPlatform'])

# Plotting the stacked bar chart
employment_platform_grouped.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='tab20')

# Adding labels and title
plt.title('Technology Adoption by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
