### Task 1: Identify Duplicate Rows

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
## Write your code here
import pandas as pd

# Load the dataset with duplicates
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv")

# Identify duplicates based on selected subset of columns
subset_cols = ['MainBranch', 'Employment', 'RemoteWork']
subset_duplicates = df[df.duplicated(subset=subset_cols, keep=False)]

print(f"Number of duplicate rows based on {subset_cols}: {subset_duplicates.shape[0]}")
print(subset_duplicates.head())

# Analyze which columns frequently contain identical values
# We'll compare row-wise similarities across all columns
grouped = subset_duplicates.groupby(subset_cols)
column_similarity = {}

# Compare each group's rows for identical values
for name, group in grouped:
    for col in df.columns:
        if col in subset_cols:
            continue  # Skip the subset columns
        # Check if all values in the column are the same within the group
        if group[col].nunique(dropna=False) == 1:
            column_similarity[col] = column_similarity.get(col, 0) + 1

# Sort by frequency of identical values across duplicates
sorted_similarity = dict(sorted(column_similarity.items(), key=lambda x: x[1], reverse=True))

print("\nColumns with most consistent values across duplicates (excluding subset columns):")
for col, freq in sorted_similarity.items():
    print(f"{col}: identical in {freq} duplicate groups")


```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset with duplicates
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv")

# Identify duplicate rows (full row duplicates)
duplicates = df[df.duplicated(keep=False)]

# 1. Bar chart: Distribution of duplicates by Country (Top 10)
country_counts = duplicates['Country'].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.barh(country_counts.index, country_counts.values, color='skyblue')
plt.title("Top 10 Countries with Most Duplicate Entries")
plt.xlabel("Number of Duplicate Rows")
plt.ylabel("Country")
plt.tight_layout()
plt.show()

# 2. Pie chart: Distribution of duplicates by Employment
employment_counts = duplicates['Employment'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(employment_counts, labels=employment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Duplicates by Employment")
plt.axis('equal')  # Ensures pie is a circle
plt.show()
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
## Write your code here
import pandas as pd

# Load dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv")

# Define columns critical to uniqueness
critical_columns = ['MainBranch', 'Employment', 'RemoteWork', 'Country', 'YearsCode']

# Check original shape
original_shape = df.shape
print(f"Original dataset shape: {original_shape}")

# Drop duplicates based on critical columns only
df_unique = df.drop_duplicates(subset=critical_columns, keep='first')

# Show new shape and how many rows were removed
new_shape = df_unique.shape
print(f"New dataset shape: {new_shape}")
print(f"Rows removed: {original_shape[0] - new_shape[0]}")
```
