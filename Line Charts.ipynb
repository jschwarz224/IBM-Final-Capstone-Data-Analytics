### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
# Install packages (only needed once)
!pip install pandas
!pip install matplotlib
!wget -O survey-data.csv https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("survey-data.csv")

# Clean data: keep only relevant columns
comp_age_df = df[['Age', 'ConvertedCompYearly']]

# Drop rows where either Age or Compensation is NaN
comp_age_df = comp_age_df.dropna()

# Group by Age and calculate the median compensation
median_comp_by_age = comp_age_df.groupby('Age')['ConvertedCompYearly'].median()

# Sort age groups logically (if needed)
age_order = ['Under 18 years old', '18-24 years old', '25-34 years old', '30-34 years old',
             '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older']
median_comp_by_age = median_comp_by_age.reindex(age_order).dropna()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', linestyle='-', color='blue')
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
# Filter for ages between 25 and 45
# We will pick relevant age groups: '25-34 years old' and '35-44 years old'
filtered_age_groups = ['25-34 years old', '30-34 years old', '35-44 years old']

# Keep only the filtered age groups
filtered_comp_by_age = median_comp_by_age.loc[median_comp_by_age.index.isin(filtered_age_groups)]

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(filtered_comp_by_age.index, filtered_comp_by_age.values, marker='o', linestyle='-', color='green')
plt.title('Median Yearly Compensation for Ages 25 to 45')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation (USD)')
plt.xticks(rotation=30)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
# Drop missing JobSatPoints_6 values
experience_job_sat = df[['Age', 'JobSatPoints_6']].dropna()

# Group by Age and calculate the median Job Satisfaction
median_job_sat_by_age = experience_job_sat.groupby('Age')['JobSatPoints_6'].median().sort_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(median_job_sat_by_age.index, median_job_sat_by_age.values, marker='o', linestyle='-', color='purple')
plt.title('Median Job Satisfaction (JobSatPoints_6) by Age Group (Experience Proxy)')
plt.xlabel('Age Group')
plt.ylabel('Median Job Satisfaction')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here
# Drop missing compensation values
experience_comp = df[['Age', 'ConvertedCompYearly']].dropna()

# Group by Age and calculate the median compensation
median_comp_by_experience = experience_comp.groupby('Age')['ConvertedCompYearly'].median().sort_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_experience.index, median_comp_by_experience.values, marker='o', linestyle='-', color='teal')
plt.title('Median Yearly Compensation by Age Group (Experience Proxy)')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here
# Drop missing job satisfaction values
experience_satisfaction = df[['Age', 'JobSatPoints_6']].dropna()

# Group by Age and calculate the median Job Satisfaction
median_satisfaction_by_experience = experience_satisfaction.groupby('Age')['JobSatPoints_6'].median().sort_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(median_satisfaction_by_experience.index, median_satisfaction_by_experience.values, marker='s', linestyle='-', color='coral')
plt.title('Median Job Satisfaction (JobSatPoints_6) by Age Group (Experience Proxy)')
plt.xlabel('Age Group')
plt.ylabel('Median Job Satisfaction (Points)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
